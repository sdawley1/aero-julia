# Vortex

A *vortex* is a region in a fluid which revolves around an axis. Vortices may form from stirred fluids and are observed in smoke rings, whirlpools, in the wake of a boat, and in the winds surrounding a tropical cyclone or tornado. 

More simply, a vortex can be thought of as a circular motion of streamlines, or, better yet, concentric circles about a given point. However, this does not necessarily imply that fluid elements themselves are rotating!

Within an irrotational vortex, the tangential velocity is constant along circular streamlines and inversely proportional to the radius, while the radial velocity remains zero. In the context of the vortex streamlines being concentric circles, these two facts should make sense. Analytically, the components of the velocity vector field of a vortex can be expressed in polar coordinates as

$$\textbf{v}(r,\theta) = \big(u(r,\theta), v(r,\theta)\big) = \left(0,~\frac{\text{constant}}{r}\right)$$

for radial and tangential components $u(r,\theta)$ and $v(r,\theta)$, respectively. Check for yourself that the vorticity of this field is zero everywhere, except at the origin of the vortex. 

The concept of *circulation* was introduced in the first module as the line integral of of the velocity vector field over a closed path. Using this definition we can easily compute the circulation of a vortex. Suppose that the closed path around the origin of the vortex we choose is a circle of radius $r$. Then, the circulation $\Gamma$ becomes

$$\Gamma = -\oint \textbf{v}\cdot d\textbf{s} = -\oint_0^{2\pi}v(r,\theta)rd\theta = -2\pi rv(r,\theta)$$

where the negative sign is used because of the convention that a negative vortex circulates counterclockwise. Thus, we have that the tangential component of the velocity vector field must be proportional to $\Gamma$ so that $\textbf{v}$ becomes

$$\textbf{v}(r,\theta) = \left(0,~\frac{\Gamma}{2\pi r}\right)$$

Recall that the stream function $\psi(r,\theta)$ is related to the radial and tangential components as $u=1/r\big(\partial\psi/\partial\theta\big)$ and $v=\partial\psi/\partial r$ so that the stream function may be determined as

$$\psi(r,\theta) = \frac{\Gamma}{2\pi}\ln r$$

For the sake of computing, this can be converted to Cartesian coordinates as

$$\psi(r,\theta) = \frac{\Gamma}{2\pi}\ln(x^2+y^2)$$

which affords the velocity vector in Cartesian coordinates:

$$\textbf{v}(x,y) = \frac{\Gamma}{2\pi}\left(\frac{y}{x^2+y^2},~-\frac{x}{x^2+y^2}\right)$$

## Visualization

We'll first set up an $x$ and $y$ axes just as we've been doing. Then, a vortex singularity is constructed using the `singularity` structure. Finally, a function to compute the velocity vector field of a vortex using the equations derived above is built and used to plot the streamlines of the vortex.

```julia
# Number of points in each direction
N = 100 

# x and y boundaries
xb = (-1.5, 1.5)
yb = (-1.5, 1.5)

# Grid of points
X = range(xb[1], xb[2], N)    
Y = range(yb[1], yb[2], N);    
```

```julia
struct singularity_point
    kind::String
    strength::Float32
    x_pos::Number
    y_pos::Number
end

# strength of the vortex
gamma = 5.0  
# location of the vortex
x_vortex, y_vortex = 0.0, 0.0

vortex = singularity_point("Vortex", gamma, x_vortex, y_vortex);
```

```julia
using CairoMakie

function vortex_field(
    source::singularity_point,
    x::Point2{Float64}
    )
    """
    source = vortex
    x = x & y positions in velocity vector grid
    """
    # Separating components
    x, y = x[1], x[2]
    
    # Velocity components
    u = (source.strength/2/pi) * (y - source.y_pos)/((x - source.x_pos)^2 + (y - source.y_pos)^2)
    v = -(source.strength/2/pi) * (x - source.x_pos)/((x - source.x_pos)^2 + (y - source.y_pos)^2)
    psi = (source.strength/4/pi) * log((x - source.x_pos)^2 + (y - source.y_pos)^2)
                
    return Point2(u, v), psi
    
end;
```

```julia
using LaTeXStrings

# Building figure
f = Figure(resolution=(600, 600))
set_theme!(
    theme_minimal(),
    font = "CM Sans Serif",
    fontsize = 18
)

# Creating first axis to plot on
ax = Axis(f[1, 1], 
    xlabel = L"$x$", 
    ylabel = L"$y$",
    title = vortex.kind
)

# Setting axis limits
xlims!(ax, (minimum(X), maximum(X)))
ylims!(ax, (minimum(Y), maximum(Y)))

# Plotting
streamplot!(x -> vortex_field(vortex, x)[1], X, Y, 
    color=:black,
    colormap=:magma, 
    arrowsize=15,
    linewidth=0.8
)
scatter!((vortex.x_pos, vortex.y_pos), color=:red)

# Displaying figure
current_figure()
```

## Vortex and a Sink

A vortex is just another type of areodynamic vector field just as sources, sinks, and freestreams are, so we can use superposition to determine the streamlines of a sink and vortex combination. 

```julia
# strength of the sink
strength_sink = -1.0
# location of the sink
x_sink, y_sink = 0.0, 0.0       

sink_field = singularity_point("Sink", strength_sink, x_sink, y_sink);
```

```julia
function sink_vortex_field(
    source1::singularity_point,
    source2::singularity_point,
    x::Point2{Float64}
    )
    """
    source1 = vortex
    source2 = sink 
    x = x & y positions in velocity vector grid
    """
    # Separating components
    x, y = x[1], x[2]
    
    # Velocity components of vortex
    source = source1
    uv = (source.strength/2/pi) * (y - source.y_pos)/((x - source.x_pos)^2 + (y - source.y_pos)^2)
    vv = -(source.strength/2/pi) * (x - source.x_pos)/((x - source.x_pos)^2 + (y - source.y_pos)^2)
    psi_v = (source.strength/4/pi) * log((x - source.x_pos)^2 + (y - source.y_pos)^2)
    
    # Velocity components of sink
    source = source2
    us = (source.strength/2/pi) * (x - source.x_pos)/((x - source.x_pos)^2 + (y - source.y_pos)^2)
    vs = (source.strength/2/pi) * (y - source.y_pos)/((x - source.x_pos)^2 + (y - source.y_pos)^2)
    psi_s = (source.strength/2/pi) * atan((y - source.y_pos), (x - source.x_pos))
                
    return Point2(uv+us, vv+vs), psi_v+psi_s
    
end;
```

```julia
# Building figure
f = Figure(resolution=(600, 600))

# Creating first axis to plot on
ax = Axis(f[1, 1], 
    xlabel = L"$x$", 
    ylabel = L"$y$",
    title = "Sink at the Center of a Vortex"
)

# Setting axis limits
xlims!(ax, (minimum(X), maximum(X)))
ylims!(ax, (minimum(Y), maximum(Y)))

# Plotting
streamplot!(x -> sink_vortex_field(vortex, sink_field, x)[1], X, Y, 
    color=:black,
    colormap=:magma, 
    arrowsize=15,
    linewidth=0.8
)
scatter!((vortex.x_pos, vortex.y_pos), color=:red)
#scatter!((sink_field.x_pos, sink_field.y_pos), color=:blue)

# Displaying figure
current_figure()
```

## Vortex-Sink Pairs

```julia
# Building figure
# Creating first axis to plot on
# Setting axis limits
# Defining theme
f = Figure(resolution=(1000, 600))

# Parameters for each of the plots
params = Dict("color" => :black, "colormap" => :magma, "linewidth" => 0.5)

# Generating random fields
fields = []
strengths = []
for i in 1:4
    pos = rand(-1:1, 2)
    s = rand(1:5)
    field = singularity_point("Type", -s, pos[1], pos[2])
    push!(fields, field)
    push!(strengths, field.strength)
end

# Scatter plots
p = reshape(fields, (2,2))

for i in 1:2
    for j in 1:2
        # Define axis
        ax = Axis(f[i,j])
        xlims!(ax, (-1.5, 1.5))
        ylims!(ax, (-1.5, 1.5))
        
        # Create streamplot
        streamplot!(ax, 
            x -> sink_vortex_field(vortex, p[j,i], x)[1], X, Y, 
            color = params["color"],
            colormap = params["colormap"], 
            linewidth = params["linewidth"]
        )
        
        # Scatter plot for vortex and sink
        scatter!(ax, (p[j,i].x_pos, p[j,i].y_pos), color=:blue)
        scatter!(ax, (vortex.x_pos, vortex.y_pos), color=:red)
    end
end

# Displaying figure
current_figure()
```

## An "Irrotational" Vortex

What exactly does it mean for a vortex to be "*irrotational*"? Surely if a vortex is present there must be rotation! 

An irrotational vector field refers to the rotation of elements *within* the field, not the field itself. Here is a very nice example (at least I think) from wikipedia of an irrotational vector field. Notice that the white lines of the balls never change their orientation. 

![](https://upload.wikimedia.org/wikipedia/commons/0/0b/Irrotational_vortex.gif)

Another way of thinking about irrotation is to consider vorticity. Remember, vorticity measures the local *angular* velocity of each element within the vector field. If the elements go in a circular path but do not spin themselves, there is no vorticity!
