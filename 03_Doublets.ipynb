# Doublets

A *doublet* can be interpreted as a combination of a source-sink pair of infinite strength kept at an infinitesimally small distance apart. Alternatively, we can imagine the source-sink pairs that have been considered up until now and pretend that we're viewing them from really far away.

We can model this situation with some simple trig. Suppose that we have a source-sink pair of strengths $\pm\Lambda$, located at $(\pm l/2, 0)$ along the $x$-axis. The position of the source versus the sink doesn't really matter due to symmetry. We've derived the stream functions for a source and a sink previously so we can determine the overall stream function of a source-sink pair by symmetry:

$$\psi=\psi_\text{source} + \psi_\text{sink} = \frac{\Lambda}{2\pi}\theta_1 + \frac{-\Lambda}{2\pi}\theta_2 = \frac{\Lambda}{2\pi}(\theta_1-\theta_2) = \frac{\Lambda}{2\pi}\Delta\theta$$

Recall that a doublet is defined by a source and a sink separated by an infinitesimal distance. So, suppose that we allow the distance $l$ between the singularities to approach zero while the magnitude of the strength increases so that the product $\Lambda l$ remains constant. In this limit, the flow pattern becomes a doublet and we define the strength of the doublet as $\kappa=\Lambda l$.

Now, we define the stream function of a doublet evaluated at some point $(x,y)$ within vector field of the fluid as

$$\psi(r,\theta) = \lim_{l\to0}\left(-\frac{\Lambda}{2\pi}d\theta\right) ~~\text{with}~~ \Lambda d=\text{constant}$$

If we consider the case where $d\theta$ is infinitesimal, we can use a little bit of trigonometry to deduce from the figure above

$$ a = l\sin\theta$$
$$ b = r - l\cos\theta $$
$$ d\theta = \frac{a}{b} = \frac{l\sin\theta}{r-l\cos\theta}$$

so that our stream function may be rewritten as

$$\psi(r,\theta) = \lim_{l\to0}\left(-\frac{\Lambda l}{2\pi}\frac{\sin\theta}{r- l\cos\theta}\right) ~~\text{with}~~ \Lambda l=\text{constant}$$

and simplifies nicely as

$$\psi(r,\theta) = -\frac{\kappa}{2\pi}\frac{\sin\theta}{r}$$

This representation is great, amazing in fact, but a more helpful representation of the stream function will be in Cartesian coordinates:

$$\psi(r,\theta) = -\frac{\kappa}{2\pi}\frac{\sin\theta}{r} = -\frac{\kappa}{2\pi}\frac{r\sin\theta}{r^2} ~\Rightarrow~\psi(x,y) = -\frac{\kappa}{2\pi}\frac{y}{x^2+y^2} $$

Recall that the stream function relates to the components of the velocity vector field $\textbf{v}(x,y)=(u,v)$ where
$u=\partial\psi/\partial y$ and $v=-\partial\psi/\partial x$ so that we can determine the components of the velocity vector field directly as

$$u=\frac{\partial\psi}{\partial y}=\frac{\partial}{\partial y}\left[-\frac{\kappa}{2\pi}\frac{y}{x^2+y^2}\right]=-\frac{\kappa}{2\pi}\frac{(x^2+y^2)-y(2y)}{(x^2+y^2)^2} = -\frac{\kappa}{2\pi}\frac{x^2-y^2}{(x^2+y^2)^2}$$

$$v=-\frac{\partial\psi}{\partial x}=-\frac{\partial}{\partial y}\left[-\frac{\kappa}{2\pi}\frac{y}{x^2+y^2}\right]=-\frac{\kappa}{2\pi}\frac{2xy}{(x^2+y^2)^2}$$

With the velocity vector for a doublet defined, we can finally get to coding! Note that the derivation above assumes that the doublet is centered at the origin. Translating the center is done easily:

$$\textbf{v}(x,y)=-\frac{\kappa}{2\pi}\left(\frac{(x-x_\text{doublet})^2-(y-y_\text{doublet})^2}{((x-x_\text{doublet})^2+(y-y_\text{doublet})^2)^2},~\frac{2(x-x_\text{doublet})(y-y_\text{doublet})}{((x-x_\text{doublet})^2+(y-y_\text{doublet})^2)^2}\right)$$


Firstly, defining some parameters:

```julia
# Number of points in each direction
N = 100 

# x and y boundaries
xb = (-1.5, 1.5)
yb = (-1.5, 1.5)

# Grid of points
X = range(xb[1], xb[2], N)    
Y = range(yb[1], yb[2], N);    
```

We'll also introduce the structures for a singularity point, as well as defining the doublet field.

```julia
struct singularity_point
    kind::String
    strength::Float32
    x_pos::Number
    y_pos::Number
end

# Doublet parameters
kappa = 3.0                        
x_doublet, y_doublet = 0.0, 0.0  
doublet = singularity_point("Doublet", kappa, x_doublet, y_doublet);
```

Similar to last time, the function `doublet_field()` will be used to compute the components of the velocity vector field and stream function for each point in the grid of the doublet.

```julia
using CairoMakie

function doublet_field(
        source::singularity_point, 
        x::Point2{Float64}
    )
    """
    source = doublet
    x = x & y positions of vector field
    """
    # Separating tuple
    x, y = x[1], x[2]
    
    # Doublet
    u = -source.strength/(2*pi) * ((x-source.x_pos)^2-(y-source.y_pos)^2)/((x-source.x_pos)^2 + (y-source.y_pos)^2)^2
    v = -source.strength/(2*pi) * (2*(x-source.x_pos)*(y-source.y_pos))/((x-source.x_pos)^2 + (y-source.y_pos)^2)^2
    
    # Superposition of outputs
    # Return point (u,v) for the streamplot
    # also return the streamline psi
    return Point(u, v)
end;
```

```julia
# Building figure
f = Figure(resolution=(600, 600))
set_theme!(
    theme_minimal(),
    font = "CM Sans Serif",
    fontsize = 18
)

# Creating axis to plot on
ax = Axis(f[1, 1], 
    xlabel = L"$x$", 
    ylabel = L"$y$",
    title = doublet.kind
)

# Setting axis limits
xlims!(ax, (minimum(X), maximum(X)))
ylims!(ax, (minimum(Y), maximum(Y)))

# Plotting
streamplot!(ax,
    x -> doublet_field(doublet, x), X, Y, 
    color=:black,
    colormap=:magma, 
    arrowsize=15,
    linewidth=0.8
)
scatter!((doublet.x_pos, doublet.y_pos), color=:red, label="Doublet")

# Displaying figure
current_figure()
```

Just as we would imagine a source-sink pair would look from far away!

## Doublet Within a Freestream

A doublet alone does not afford much information about how it may be used to represent practical flow patterns in aerodynamics. In fact, it look almost useless. Though, the fluid flow of a doublet within a freestream affords a very interesting pattern. Let's visualize it!

We'll start by defining a freestream structure and building a freestream with a speed of unity. Recall from previous modules that the stream function of uniform flow in the horizontal direction is given simply by $\psi(x,y)=U_\infty y$ so that the overall stream function is given by superposition:

$$ \psi(x,y)=\psi_\text{freestream} + \psi_\text{doublet} = U_\infty y - \frac{\kappa}{2\pi}\frac{y}{x^2+y^2}$$

```julia
struct freestream
    speed::Number
    # angle in radians
    angle::Number
end

fs = freestream(1.0, 0);
```

Now we define the function `doublet_horizontal_freestream()` to compute the velocity vector field of a doublet within a freestream, using the doublet and freestream structures we've already created. 

```julia
function doublet_horizontal_freestream(
        source::singularity_point, 
        fs::freestream, 
        x::Point2{Float64}
    )
    """
    source = singularity point
    fs = freestream
    x = x & y positions of vector field
    """
    # Separating tuple
    x, y = x[1], x[2]
    
    # Doublet
    ud = -source.strength/(2*pi) * ((x-source.x_pos)^2-(y-source.y_pos)^2)/((x-source.x_pos)^2 + (y-source.y_pos)^2)^2
    vd = -source.strength/(2*pi) * (2*(x-source.x_pos)*(y-source.y_pos))/((x-source.x_pos)^2 + (y-source.y_pos)^2)^2
    psi_d = -source.strength /(2*pi) * (y - source.y_pos) / ((x - source.x_pos)^2 + (y-source.y_pos)^2)
    
    # Freestream 
    uf = fs.speed
    vf = 0
    psi_f = fs.speed * y
    
    # Superposition of outputs
    # Return point (u,v) for the streamplot
    # also return the streamline psi
    return Point(ud+uf, vd+vf), psi_d+psi_f
end;
```

```julia
using LaTeXStrings

# Building figure
f = Figure(resolution=(600, 600))
set_theme!(
    theme_minimal(),
    font = "CM Sans Serif",
    fontsize = 18
)
# Axis to plot on
ax = Axis(
    f[1,1],
    title = "Source in a Freestream", 
    xlabel = L"$x$",
    ylabel = L"$y$"
)
xlims!(minimum(X), maximum(X))
ylims!(minimum(Y), maximum(Y))


# Plotting
streamplot!(ax,
    x -> doublet_horizontal_freestream(doublet, fs, x)[1], X, Y, 
    color=:black,
    colormap=:magma, 
    arrowsize=15,
    linewidth=0.8
)

# Contour 
# Levels of the contour plot are determined below
P = [doublet_horizontal_freestream(doublet, fs, Point2(x, y))[2] for x in X, y in Y]
contour!(X, Y, P, 
    levels = [0],
    color = :red, 
    linewidth = 3
)

# Scatter
scatter!((doublet.x_pos, doublet.y_pos), color=:red, label="Doublet")
#sc_source = scatter!(ax, (source_field.x_pos, source_field.y_pos), color=:red)
#sc_stagnation = scatter!(ax, (x_stag, y_stag), color=:cyan)

# Legend
#Legend(f[1,2], [sc_source, sc_stagnation], ["Source", "Stagnation"])

# Displaying figure
current_figure()
```

The contour plot reveals that the *dividing streamline* can be represented as a cylinder centered at the doublet. All streamlines remaining outside of this cylinder originate from the freestream while all streamlines within the cylinder come from the doublet. If we further assume that this cylinder is a solid object we can simply ignore the streamlines within the circle. This assumption will be very useful in the future. 

## The Streamlines of a Doublet in Polar Coordinates

Converting to polar coordinates affords the following expression for the streamlines of a doublet in a freestream of uniform flow:

$$ \psi(r,\theta) = U_\infty r\sin\theta - \frac{\kappa}{2\pi}\frac{\sin\theta}{r}$$

The conversion to polar coordinates is to give us an idea of where we'll find the dividing streamlines, that is, where $\psi=0$. In the first equation expression in Cartesian coordinates, it's not difficult to see that one condition required for $\psi=0$ is that $y=0$. Or, if the center of the doublet is shifted, $y=y_\text{doublet}$. Similarly, in polar coordinates, we can see that one of the conditions for $\psi=0$ is that $\theta=0$ or $\theta=\pi$ (where $\theta$ is restricted to the interval of $[0,2\pi]$). With these conditions in mind, we can now determine where the velocity vector is zero, i.e., there exist stagnation points.

Firstly, the $x$ component of the velocity vector field, $u(x,y)$. An expression for $\psi$ is given above so that this component of the velocity can be determined via differentiation:

$$ 0 = u = \frac{\partial\psi}{\partial y} = \frac{\partial}{\partial y}\left[U_\infty y - \frac{\kappa}{2\pi}\frac{y}{x^2+y^2}\right] = U_\infty - \frac{\kappa}{2\pi}\frac{x^2-y^2}{(x^2+y^2)^2} = U_\infty-\frac{\kappa}{2\pi}\frac{x^2}{(x^2)^2} ~\Rightarrow~ x^2 = \frac{\kappa}{2\pi U_\infty}$$

Recall that we can convert between Cartesian and polar coordinates using the identity $x=r\cos\theta$. However, we've already determined that one of the conditions of our system is that $\theta=0$ or $\pi$ so that $x=r$. Thus, the radius of the cylinder formed by a doublet in a freestream is given by

$$ R = \sqrt{\frac{\kappa}{2\pi U_\infty}}$$

The stream function of the doublet in polar coordinates has been mentioned a few times already. Though, up until this point we've only studied the velocity vector components in Cartesian coordinates. Now, we'll consider the velocity vector components of the fluid in polar coordinates. 

Just as we can relate $u$ and $v$ of the velocity vector to the stream function in Cartesian coordinates by differentation, we can do the same in polar coordinates, with some slight modifications due to the chain rule:

$$u = \frac{\partial\psi}{\partial r},\quad v = -\frac{1}{r}\frac{\partial\psi}{\partial\theta}$$

Above, the velocity vector field will take the form $\textbf{v}(r,\theta)=(u(r,\theta), v(r,\theta))$ where $u$ and $v$ become the radial and tangential components of the velocity vector, respectively. Determining expressions for these components requires a small bit of differentiation:

$$u(r,\theta) =\frac{1}{r}\frac{\partial\psi}{\partial\theta} = \frac{1}{r}\frac{\partial}{\partial\theta}\left[U_\infty r\sin\theta - \frac{\kappa}{2\pi}\frac{\sin\theta}{r}\right] = \left(U_\infty - \frac{\kappa}{2\pi r^2}\right)\cos\theta$$

$$ v(r,\theta) = -\frac{\partial\psi}{\partial r} = -\frac{\partial}{\partial r}\left[U_\infty r\sin\theta - \frac{\kappa}{2\pi}\frac{\sin\theta}{r}\right] = -\left(U_\infty + \frac{\kappa}{2\pi r^2}\right)\sin\theta $$

What does this tell us about the components of the velocity vector field? If you're familiar with some aspects of symmetry, you might immediately see that from this representation we know that *the components of the velocity vector field along the dividing streamline are everywhere orthogonal*. To make this more clear, consider the integral of the product $uv$ over all angles $\theta$:

$$\int_\text{all $\theta$}uv~d\theta = -\int_0^{2\pi}\left(U_\infty - \frac{\kappa}{2\pi r^2}\right)\cos\theta\left(U_\infty + \frac{\kappa}{2\pi r^2}\right)\sin\theta d\theta = -\left(U_\infty^2 - \left(\frac{\kappa}{2\pi r^2}\right)^2\right)\int_0^{2\pi}\sin\theta\cos\theta d\theta = 0$$

As we can see above, sine and cosine are orthogonal functions. So, the components of the velocity vector field are orthogonal. Intuitively, this should make sense because the dividing streamline is a circle.

## Bernoulli's Equation and the Pressure Coefficient

Recall from the previous module how we may write the pressure coefficient, $C_p$, for a velocity vector $\textbf{v}$ and speed of a freestream $U_\infty$.

$$C_p = 1 - \left(\frac{|\textbf{v}|}{U_\infty}\right)^2$$

We've previously determined that the radius of the cylinder defined by the dividing streamlines is given by $R^2=\kappa/(2\pi U_\infty)$. So, *at the surface of the cylinder*, we find that the speed of the freestream must be

$$U_\infty = \frac{\kappa}{2\pi R^2}$$

Now, we can determine an alternate representation of the pressure coefficient. A moment ago we determined the velocity vector components as a function of $r$ and $\theta$. So, we can now determine what that these components are at the surface of the cylinder:

$$u(R,\theta) = \left(U_\infty - \frac{\kappa}{2\pi R^2}\right)\cos\theta = \left(\frac{\kappa}{2\pi R^2} - \frac{\kappa}{2\pi R^2}\right)\cos\theta = 0$$

$$ v(R,\theta) = -\left(U_\infty + \frac{\kappa}{2\pi R^2}\right)\sin\theta = -\left(\frac{\kappa}{2\pi R^2} + \frac{\kappa}{2\pi R^2}\right)\sin\theta = -\frac{\kappa}{\pi R^2}\sin\theta$$

A-ha! At the surface of the cylinder only the tangential component of the velocity vector contributes to the magnitude of the velocity, that is, there is no radial component of the velocity vector. So, that magnitude of the velocity vector is equivalent to the magnitude of the tangential component of the velocity. Using this result, and our expression for the freestream velocity in terms of the radius of the cylinder, in our definition of the pressure coefficient allows us to find an elegant expression for $C_p$:

$$C_p = 1 - \left(\frac{|\textbf{v}|}{U_\infty}\right)^2 = 1 - \left(\frac{v}{U_\infty}\right)^2 = 1 - \left(\frac{-\frac{\kappa}{\pi R^2}\sin\theta}{\frac{\kappa}{2\pi R^2}}\right)^2 = 1-4\sin^2\theta$$

The pressure coefficient as a function of the angle $\theta$ is plotted below.

```julia
using LaTeXStrings

#Building figure
f = Figure(resolution=(400, 400))

# Axis and limits
ax = Axis(
    f[1,1],
    title = "Pressure Coefficient as Function \nof Polar Angle", 
    xlabel = L"$\theta$",
    xlabelsize=20,
    ylabel = "Pressure Coefficient",
    ylabelsize=20
)

# X axis and pressure coefficient
# X axis represents polar angle, Y axis
X = range(0, 2*pi, 100)
Y = [1-4*(sin(x))^2 for x in X]

lines!(ax, X, Y, color=:red, linewidth=3)

# Displaying figure
current_figure()
```

Now, the pressure coefficient around the doublet is plotted on a heatmap.

```julia
using LaTeXStrings

# Number of points in each direction
N = 100 

# x and y boundaries
xb = (-1.5, 1.5)
yb = (-1.5, 1.5)

# Grid of points
X = range(xb[1], xb[2], N)    
Y = range(yb[1], yb[2], N);    

#Building figure
f = Figure(resolution=(700, 600))
set_theme!(
    theme_minimal(),
    font = "CM Sans Serif",
    fontsize = 18
)
# Axis and limits
ax = Axis(
    f[1,1],
    title = "Pressure Coefficient Around Doublet", 
    xlabel = L"$x$",
    ylabel = L"$y$"
)
xlims!(minimum(X), maximum(X))
ylims!(minimum(Y), maximum(Y))

# Plotting
# First, grab velocity vector (V) and stream function (P) from doublet in freestream
V = [doublet_horizontal_freestream(doublet, fs, Point2(x, y))[1] for x in X, y in Y]
P = [doublet_horizontal_freestream(doublet, fs, Point2(x, y))[2] for x in X, y in Y]

# Heat map of pressure coefficients
cp = [1.0 - (sqrt(i[1]^2 + i[2]^2) / fs.speed)^2 for i in V]
cp = [1.5^i for i in cp] # Increasing contrast
heatmap!(ax, 
    X, Y, cp,
    colormap = :magma, 
)

# Contour 
contour!(X, Y, P, 
    levels = [0],
    color = :red, 
    linewidth = 3
)

# Colorbar
cb_limits = (log(minimum(cp), 1.4), log(maximum(cp), 1.4))
cb = Colorbar(f[1, 2], colormap=:magma, limits=cb_limits, label="Pressure Coefficient")

# Displaying figure
current_figure()
```
